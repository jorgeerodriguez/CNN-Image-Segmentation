```python
def copy_and_rename():	
    source_folder = "./archive/train/nike"
    dest_folder = "./images"
    for count, filename in enumerate(os.listdir(source_folder)):
        new_filename = f"nike_{filename.split(".")[0]}.jpg"
        print(filename.split(".")[0])
        src = f"{source_folder}/{filename}" # foldername/filename, if .py file is outside folder
        print(src)
        dst = f"{dest_folder}/{new_filename}" # foldername/filename, if .py file is outside folder
        print(dst)
        os.rename(src, dst)
		#dst =f"{folder}/{dst}"
		
		# rename() function will
		# rename all the files
		#os.rename(src, dst)
```
